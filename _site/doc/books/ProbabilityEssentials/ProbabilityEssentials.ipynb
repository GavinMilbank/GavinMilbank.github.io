# 2. Axioms
* Let $\Omega$ be an abstract event space with no special structure.
* $2^{\Omega}$ denotes all subsets of $\Omega$ including the empty set $\emptyset$ 
* Let $\mathcal{A}$ be subset of $2^{\Omega}$, consider following properties:
	1. emptyset is member : $\emptyset \in \mathcal{A} $
	2. closed under compliment: if $A \in \mathcal{A}$ then $A^{c} \in \mathcal{A}$
	3. closed under finite unions and intersestions : $ A_{n} \in \mathcal{A}$ then $\cup_{n}^{N} A_{n} \in \mathcal{A}$ and  $\cap_{n}^{N} A_{n}$
	4. closed under countable unions and intersections : $ A_{n} \in \mathcal{A}$ then $\cup_{n}^{\infty} A_{n} \in \mathcal{A}$ and  $\cap_{n}^{\infty} A_{n}$

#### Definition 2.1 - $\sigma$-algebras
* $\mathcal{A}$ is an algebra if it satisfies 1,2 and 3 above.
* $\mathcal{A}$ is an $\sigma$-algebra if it satisfies 1,2,3 and 4 above.


#### Definition 2.2 - $\sigma$-algebras
* If $\mathcal{C} \subset 2^{\Omega}$ then $\sigma(\mathcal{C})$ is the smallest $\sigma$-algebra generated by $\mathcal{C}$  
* $\sigma(\mathcal{C})$ contains every set (or state of the world) expressible by logical statements on $\mathcal{C}$

#### Theorem 2.1 - Borel $\sigma$-algebra
* The Borel $\sigma$-algebra of $\mathcal{R}$ is generated by intervals of the form $(-\infty,a])$ where $a \in \mathcal{Q}$(rationals)
* *Proof*:
	* $\mathcal{C}$ = all open intervals, Borel $\sigma$-algebra $\mathcal{B}$ is $\sigma{\mathcal(C)}$
	* $\mathcal{D}$ = all open intervals from intervals of form $(-\infty,a]$
	* $a_{n}$ is sequenced of rationals decreasing strictly to a 
	* $b_{n}$ is sequenced of rationals increasing strictly to b
	* $\mathcal{C} \subset \sigma(\mathcal{D})$ because $(a,b) = \cup_{n}^{\infty}(a_{n},b_{n}] =  \cup_{n}^{\infty}((-\infty,b_{n}] \cap (-\infty,a_{n}]^{c})) $ thus $\sigma(\mathcal{C}) \subset \sigma(\mathcal{D})$
	* $\mathcal{B} = \sigma(\mathcal{C}) \subset \sigma(\mathcal{D}) \subset \mathcal{B} \Rightarrow \sigma(\mathcal{D}) = \mathcal{B}$

#### Definition 2.3 - Probability measure
* A probability measure defined on $\sigma$-algebra $\sigma$-algebra $\mathcal{A}$ of $\Omega$ a is a function $
P : \mathcal{A} \rightarrow [0,1] $ such that:
    1. $P(\Omega)$ = 1
    2. Countable additivity for pairwaise disjoint sequences $A_{n}$: $P(\cup_{n}^{\infty}A_{n}) = \sum_{n}^{\infty} P(A_{n})$
    
#### Theorem 2.2 - Probability measures are additive
If P is a probability measure on $(\Omega,\mathcal{A})$, then
1. $P(\emptyset) = 0$
2. P is additive

#### Theorem 2.3 - Implications of countable additivity
If P is an *additive* probability measure, then the following are equivalent:
1. Defintion 2.3.2. i.e. *countable* additivity for pairwaise disjoint sequences
2. $A_{n} \in \mathcal{A}$ and $A_{n} \downarrow \emptyset$, then $P(A_{n}) \downarrow 0$
3. $A_{n} \in \mathcal{A}$ and $A_{n} \downarrow A$, then $P(A_{n}) \downarrow P(A)$
4. $A_{n} \in \mathcal{A}$ and $A_{n} \uparrow \Omega$, then $P(A_{n}) \uparrow 1$
5. $A_{n} \in \mathcal{A}$ and $A_{n} \uparrow A$, then $P(A_{n}) \uparrow P(A)$
*Proof*: use countable additivity

*Indicator function* on $A \in 2^{\Omega}$ defined by: $1_{A}(\omega)$ = 1 if $\omega \in A$ else 0

$A_{n} \in \mathcal{A}$ *converges* to $A$, written $A_n \rightarrow A$, if $lim_{n\to\infty} 1_{A_{n}}(\omega) = 1_{A}(\omega)$ for all $\omega \in \Omega$

#### Theorem 2.4 - Convergence in probability measure space
If $A_{n}\rightarrow A$ in $\mathcal{A}$ then $A \in \mathcal{A}$ and $\lim_{n\to\infty} P(A_{n}) = P(A)$.

*Proof*
- define [limits of sequence of extrema](https://en.wikipedia.org/wiki/Limit_superior_and_limit_inferior#Interpretation)
    * $\limsup_{n \to \infty}A_{n} = \cap^{\infty}_{n=1} \cup_{m \geq n} A_{m}$
    * $\liminf_{n \to \infty}A_{n} = \cup^{\infty}_{n=1} \cap_{m \geq n} A_{m}$
- $\mathcal{A}$ is a $\sigma$-algebra so $\limsup_{n \to \infty}A_{n} \in \mathcal{A}$ and $\liminf_{n \to \infty}A_{n} \in \mathcal{A}$
- $A_{n}\rightarrow A$ $\Leftrightarrow$ $\lim_{n \to \infty}1_{A} = 1_{A}$,all $\omega$ $\Leftrightarrow$ $A = \limsup_{n \to \infty}A_{n} = \liminf_{n \to \infty}A_{n}$  $\Leftrightarrow$  $A \in \mathcal{A}$
- $B_{n}=\cap_{m \geq n} A_{m}$ and $C_{n}=\cup_{m \geq n} A_{m}$, then $B_{n} \uparrow A$ and $C_{n} \downarrow A$, so $\lim{n\to\infty}P(B_{n})=\lim{n\to\infty}P(C_{n})=P(A)$ by theorem 2.3.
- but $B_{n} \subset A_{n} \subset C_{n}$ so $P(B_{n}) \leq P(A_{n}) \leq P(C_{n})$ so $\lim_{n\to\infty}P(A_{n})=P(A)$ 

# 3. Conditional Probability and Independence
* Independence $ A \perp B  \Leftrightarrow P(A \cap B) = P(A)P(B) $
* Conditioning $ P(A|B) = P(A \cap B)/P(B) $
* Partition $ P(A) = \sum_{n} P(A|E_{n})P(E_{n}) $
* Bayes $ P(E_{n}|A) = P(A|E_{n}) P(E_{n})/\sum_{m} P(A|E_{n})P(E_{n}) = P(A \cap E_{n})/P(A)$

# 4. Probability on Countable Space

####  4.a. Hypergeometric Distribution (*opinion polls*)
* N white balls and M black balls, draw n < N+M, without replacement => X white balls and n-X black balls
* $\Omega$ is family of all subsets with n points, the cardinality of which $ |\Omega| = \binom{N+M}{n} $
* where $\binom pq = \frac{q!}{p!(q-p)!}$ , choose p from q (unordered).     
* $X^{-1}({x})$ contains $ \binom{N}{x} \binom{M}{n-x} $ points if $x\leq N$ and $n-x\leq M$ else is empty
* $ P(X=x) = \frac{\binom{N}{x} \binom{M}{n-x}}{\binom{N+M}{n}} $
* binomial as limit of hypergeometric

* $ P(X=x) \Rightarrow \binom ab  p^{x} (1-p)^{n-x} $
